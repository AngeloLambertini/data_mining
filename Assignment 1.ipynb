### Problem 2.5.
Using the concept of overfitting, explain why when a model is fit to training data, zero error with those data is not necessarily good.

Seeing a result of 0 errors when a model is fit to training data is not necessarily good because having such result usually ends up in a problem of overfitting. The overfitting happens when the model which is being trained fits the training data so perfectly that it includes all the variation of the data, modeling also that part of variation that is only chance variation and does not give any help in understanding underlying patterns. An overfitting model will perfeclty describe our training data but will not be useful in working with new data and making predictions, which is usually the main purpose when developing a model.
