The dataset on American college and university rankings contains information on 1302 American colleges and universities
offering an undergraduate program. For each university, there are 17 measurements that include
continuous measurements (such as tuition and graduation rate) and categorical measurements
(such as location by state and whether it is a private or a public school).
* a. Remove all categorical variables. Then remove all records with missing numerical
measurements from the dataset.
* b. Conduct a principal components analysis on the cleaned data and comment on the results.
Should the data be normalized? Discuss what characterizes the components you consider key.

```python
uni_df = pd.read_csv('Universities.csv')
```

```python
uni_df.head(5)
```

```python
# Delete cathegorical variables
DeleteList=['College Name','State','Public (1)/ Private (2)']
uni_df=uni_df.drop(DeleteList, axis=1)
```

```python
uni_df.head(5)
```

```python
uni_df.isnull().values.any()
```

```python
print('Number of missing values :\n', uni_df.isnull().sum())
```

```python
uni_df.describe()
```

```python
uni_df2 = uni_df.dropna(how='any')
```

```python
uni_df2.describe()
```

```python
pcs = PCA()
pcs.fit(uni_df2.dropna(axis=0))
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
'Proportion of variance': pcs.explained_variance_ratio_,
'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns=pcsSummary_df.columns,
index=uni_df2.columns)
pcsComponents_df.iloc[:,:14]
```

4b) Running the PCA anaysis, we can see that PC1 alone should account for over 56% of the total variability, and adding up the frist three principal components we should be grouping more than 96% of it. The 99% of the variability should be reached with PC1 - PC7 and the last 5 principal components wouldn't add any information about the variability in the data. 
The key components would be PC1 and PC2. PC1 mostly measures the balance between application received and full-time undergraduate students (large positive wheights) vs in state tuition and out of state tuition (large negative weights); high scores on PC1 mean that the university has a high number of applications and full time students and low tuition fees for both domestic and international students. PC2 is heavily characterized by the number of application received, the number of students, and the out of state tuition (all with large positive weights), universities with high scores will have a large numer of applicants and students and high international students fees.

All this being said, the variables summary shows large differencies in the variables scales, so <b>the data should be normalized before conducting the PC Analysis to gain accurate insights</b>, in the analysis conducted above, the weights of the different variables are affected by their corresponding scales, meaning that the wheight of variables with wider range and mean will be overestimated.
